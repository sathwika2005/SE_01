```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_model(time):
    # Hard coded coefficients for quadratic equation
    a = 0.1
    b = -1
    c = 30
    # Quadratic equation representing temperature change
    temperature = a * (time ** 2) + b * time + c
    return temperature, a, b, c

def main():
    # Time values from 0 to 10
    time_values = np.linspace(0, 10, 50)

    # Calculate temperature using hardcoded coefficients
    temp_hardcoded, a, b, c = quadratic_model(time_values)

    # Plot results
    plt.plot(time_values, temp_hardcoded, label=f"Hard-coded Coefficients: a={a}, b={b}, c={c}")
    plt.xlabel("Time")
    plt.ylabel("Temperature")
    plt.legend()
    plt.title("Weather Modeling with Quadratic Equation (Hard-coded Coefficients)")
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def read_coefficients(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

        # Initialize lists to store coefficients
        a_values = []
        b_values = []
        c_values = []

        # Read coefficients from each line
        for line in lines:
            coeffs = line.strip().split()
            if len(coeffs) == 3:
                a_values.append(float(coeffs[0]))
                b_values.append(float(coeffs[1]))
                c_values.append(float(coeffs[2]))

        return a_values, b_values, c_values

def quadratic_model(a, b, c, time):
    # Calculate temperature for multiple sets of coefficients
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    filename = 'qm.txt'
    a_values, b_values, c_values = read_coefficients(filename)

    time_values = np.linspace(0, 10, 50)

    # Plotting multiple curves
    for a, b, c in zip(a_values, b_values, c_values):
        temp_file = quadratic_model(a, b, c, time_values)
        plt.plot(time_values, temp_file, label=f"a={a}, b={b}, c={c}")

    plt.xlabel("Time")
    plt.ylabel("Temperature")
    plt.legend()
    plt.title("Weather Modeling with Quadratic Equations")
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_model(time, a, b, c):
    # Quadratic equation representing temperature change
    temperature = a * (time ** 2) + b * time + c
    return temperature

def get_coefficients():
    # Prompt the user for coefficients
    a = float(input("Enter the value of a: "))
    b = float(input("Enter the value of b: "))
    c = float(input("Enter the value of c: "))
    return a, b, c

def main():
    # Time values from 0 to 10
    time_values = np.linspace(0, 10, 50)

    # Hardcoded coefficients
    a_hardcoded = 15
    b_hardcoded = 45
    c_hardcoded = 60
    temp_hardcoded = quadratic_model(time_values, a_hardcoded, b_hardcoded, c_hardcoded)

    # Get coefficients from user input
    a_keyboard, b_keyboard, c_keyboard = get_coefficients()
    temp_keyboard = quadratic_model(time_values, a_keyboard, b_keyboard, c_keyboard)

    # Plot results
    plt.plot(time_values, temp_hardcoded, label=f"Hard-coded Coefficients: a={a_hardcoded}, b={b_hardcoded}, c={c_hardcoded}")
    plt.plot(time_values, temp_keyboard, label=f"Keyboard Input Coefficients: a={a_keyboard}, b={b_keyboard}, c={c_keyboard}")
    plt.xlabel("Time")
    plt.ylabel("Temperature")
    plt.legend()
    plt.title("Weather Modeling with Quadratic Equation")
    plt.show()

if __name__ == "__main__":
    main()
```
